## Aufgabe 1 – Bibliotheken importieren

Importieren Sie:
- `load_iris` aus `sklearn.datasets`
- `pandas` als `pd`
- `matplotlib.pyplot` als `plt`


```python
# Musterlösung Aufgabe 1
from sklearn.datasets import load_iris
import pandas as pd
import matplotlib.pyplot as plt
```

## Aufgabe 2 – Iris-Datensatz laden

Laden Sie den Datensatz in eine Variable `iris`.


```python
# Musterlösung Aufgabe 2
iris = load_iris()
```

## Aufgabe 3 – X und y erzeugen

Erzeugen Sie:
- `X` als Feature-Matrix
- `y` als Zielvariable


```python
# Musterlösung Aufgabe 3
X = iris.data
y = iris.target
```

## Aufgabe 4 – DataFrame erzeugen

Erstellen Sie einen DataFrame aus `X` mit den Spaltennamen aus `iris.feature_names`.  
Fügen Sie zusätzlich eine Spalte `species` mit den Klassennamen ein.


```python
# Musterlösung Aufgabe 4
df = pd.DataFrame(X, columns=iris.feature_names)
df["species"] = [iris.target_names[i] for i in y]

df.head()
```

## Aufgabe 5 – Erste Analyse

Geben Sie aus:
- `X.shape`
- `y.shape`
- die ersten 5 Zeilen des DataFrames
- die Klassennamen


```python
# Musterlösung Aufgabe 5
print("X.shape =", X.shape)
print("y.shape =", y.shape)
print()
print(df.head())
print()
print("Klassennamen:", iris.target_names)
```

## Aufgabe 6 – Visualisierung

Erstellen Sie ein Streudiagramm für die ersten beiden Merkmale.  
Beschriften Sie die Achsen sinnvoll und fügen Sie eine Farblegende hinzu.


```python
# Musterlösung Aufgabe 6
plt.figure(figsize=(7, 5))
scatter = plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis')
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
plt.title("Iris-Daten: erstes gegen zweites Merkmal")
plt.colorbar(scatter, label='Klasse (0=setosa, 1=versicolor, 2=virginica)')
plt.grid(True)
plt.show()
```

## Musterantworten zur Reflexion

- **Sample:** Ein einzelner Datenpunkt bzw. eine Beobachtung im Datensatz.
- **Feature:** Ein Merkmal, das ein Sample beschreibt.
- **Label:** Der Zielwert bzw. die Klasse, die vorhergesagt werden soll.
- **Warum Visualisierungen hilfreich sind:** Sie machen Muster, Trennbarkeit, Ausreißer und Zusammenhänge schneller sichtbar als reine Tabellen.
