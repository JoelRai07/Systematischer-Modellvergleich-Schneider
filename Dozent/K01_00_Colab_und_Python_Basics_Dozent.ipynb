## Abschnitt 2 – Listen und Schleifen

Listen sind geordnete Sammlungen von Werten.  
Mit einer Schleife können wir über alle Elemente einer Liste nacheinander iterieren.


```python
zahlen = [2, 4, 6, 8, 10]
print("Liste:", zahlen)
print("Erstes Element:", zahlen[0])
print("Anzahl Elemente:", len(zahlen))
```

```python
for z in zahlen:
    print("Quadrat von", z, "=", z**2)
```

Ein **DataFrame** ist eine tabellarische Datenstruktur mit Zeilen und Spalten.

## Abschnitt 5 – Erste Visualisierung

Wir erstellen ein einfaches Liniendiagramm.  
Später werden wir im Kurs häufiger Streudiagramme verwenden.


```python
plt.figure(figsize=(6, 4))
plt.plot(df["Lernzeit_Stunden"], df["Punktzahl"], marker="o")
plt.xlabel("Lernzeit in Stunden")
plt.ylabel("Punktzahl")
plt.title("Lernzeit und Punktzahl")
plt.grid(True)
plt.show()
```

### Musterlösung zu den Mini-Aufgaben

**Aufgabe 1:** Wenn die Werte in `Lernzeit_Stunden` verändert werden, verschieben sich die Punkte bzw. die Linie auf der x-Achse entsprechend. Das Diagramm aktualisiert sich beim erneuten Ausführen der Plot-Zelle.



```python
df = pd.DataFrame({
    "Lernzeit_Stunden": [2, 3, 7, 18, 20],
    "Punktzahl": [50, 58, 68, 80, 92]
})


plt.figure(figsize=(6, 4))
plt.plot(df["Lernzeit_Stunden"], df["Punktzahl"], marker="o")
plt.xlabel("Lernzeit in Stunden")
plt.ylabel("Punktzahl")
plt.title("Lernzeit und Punktzahl")
plt.grid(True)
plt.show()
```

### Musterlösung zu den Mini-Aufgaben


**Aufgabe 2:** Eine neue Zeile kann z. B. so hinzugefügt werden (siehe Code-Zelle unten).




```python
# Musterlösung Aufgabe 2: Neue Zeile hinzufügen
df_neu = df.copy()
df_neu.loc[len(df_neu)] = [12, 98]
print(df_neu)


plt.figure(figsize=(6, 4))
plt.plot(df_neu["Lernzeit_Stunden"], df_neu["Punktzahl"], marker="o")
plt.xlabel("Lernzeit in Stunden")
plt.ylabel("Punktzahl")
plt.title("Lernzeit und Punktzahl")
plt.grid(True)
plt.show()
```

### Musterlösung zu den Mini-Aufgaben

**Aufgabe 3:** Für ein Streudiagramm wird `plt.scatter()` statt `plt.plot()` verwendet (siehe Code-Zelle unten).


`plt.scatter(df["Lernzeit_Stunden"], df["Punktzahl"])`: Dies ist der Kern des Streudiagramms. Es wird ein Punkt für jedes Datenpaar gezeichnet, wobei die 'Lernzeit_Stunden' auf der X-Achse und die 'Punktzahl' auf der Y-Achse liegen.



```python
# Musterlösung Aufgabe 3: Streudiagramm
plt.figure(figsize=(6, 4))
plt.scatter(df["Lernzeit_Stunden"], df["Punktzahl"])
plt.xlabel("Lernzeit in Stunden")
plt.ylabel("Punktzahl")
plt.title("Streudiagramm: Lernzeit und Punktzahl")
plt.grid(True)
plt.show()
```

### Musterlösung zu den Mini-Aufgaben

**Verständnisfrage 1 – Liste vs. NumPy-Array:**  
Eine Python-Liste ist eine allgemeine Datenstruktur, die beliebige Typen mischen kann (z. B. `[1, 'a', True]`). Ein NumPy-Array ist auf einen einheitlichen Datentyp spezialisiert und unterstützt direkte Vektoroperationen wie `arr * 2` oder `arr + 10`, ohne eine Schleife zu benötigen. Das macht NumPy für numerische Berechnungen deutlich effizienter.

**Verständnisfrage 2 – DataFrame vs. Liste:**  
Ein DataFrame hat benannte Spalten, ermöglicht einfache Filterung, Aggregation und Visualisierung und ähnelt einer Excel-Tabelle. Bei mehreren Merkmalen (Spalten) wird ein DataFrame schnell übersichtlicher und mächtiger als parallele Listen.
