# Lineare Regression anhand eines einfachen Beispieles

Modell: `LinearRegression()`

Daten: `koerpergroesse vs. ringfingerumfang`

Metrik: `r2_score`

# Daten visualisieren

```python
import matplotlib.pyplot as plt

# Daten
personen = ["Anna", "Bernd", "Catharina", "David", "Edith",
            "Frank", "Gertrud", "Hans", "Irma", "Jochen"]
koerpergroesse = [153.3, 158.9, 160.8, 179.6, 156.6,
                  165.1, 165.9, 156.7, 167.8, 160.8]
ringfingerumfang = [47.1, 46.8, 49.3, 53.2, 47.7,
                    49.0, 50.6, 47.1, 51.7, 47.8]

# Plot
plt.figure(figsize=(10, 6))
plt.scatter(koerpergroesse, ringfingerumfang, color='blue')

# mit Beschriftung
for i, name in enumerate(personen):
    plt.text(koerpergroesse[i] + 0.3, ringfingerumfang[i], name, fontsize=9)

plt.title("Zusammenhang zwischen Körpergröße und Ringfingerumfang")
plt.xlabel("Körpergröße (cm)")
plt.ylabel("Ringfingerumfang (mm)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

# Lineare Regression durchführen

```python

from sklearn.linear_model import LinearRegression
import numpy as np


# Umformen der Daten für scikit-learn (2D-Array)
X = np.array(koerpergroesse).reshape(-1, 1)
y = np.array(ringfingerumfang)

# Regressionsmodell erstellen und trainieren
modell = LinearRegression()
modell.fit(X, y)

# Vorhersagewerte für die Regressionslinie
x_linspace = np.linspace(min(koerpergroesse), max(koerpergroesse), 100).reshape(-1, 1)
y_pred = modell.predict(x_linspace)

# Plot
plt.figure(figsize=(10, 6))
plt.scatter(koerpergroesse, ringfingerumfang, color='blue', label='Datenpunkte')

# Namen eintragen
for i, name in enumerate(personen):
    plt.text(koerpergroesse[i] + 0.3, ringfingerumfang[i], name, fontsize=9)

# Regressionslinie
plt.plot(x_linspace, y_pred, color='red', linestyle='--', label='Regressionslinie')

# Achsenbeschriftung und Titel
plt.title("Zusammenhang zwischen Körpergröße und Ringfingerumfang")
plt.xlabel("Körpergröße (cm)")
plt.ylabel("Ringfingerumfang (mm)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

# Modell Koeffizienten ausgeben

```python
# Koeffizienten
a = modell.coef_[0]
b = modell.intercept_

print(f"y = {a:.4f} * x + {b:.4f}")
```

# Bestimmtheitsmaß berechnen

```python
from sklearn.metrics import r2_score


# Vorhersage und R²-Berechnung
y_vorhersage = modell.predict(X)
r2 = r2_score(y, y_vorhersage)

print(f"Bestimmtheitsmaß (R²): {r2:.4f}")
```

# Lessons Learned

1.  **Datenvisualisierung zur Mustererkennung**:
Durch die Erstellung eines Scatter-Plots konnten wir den Zusammenhang zwischen Körpergröße und Ringfingerumfang visuell erfassen und einzelne Datenpunkte (Personen) direkt im Diagramm identifizieren. Dies ist ein erster wichtiger Schritt, um Daten zu verstehen.
---

2.  **Anwendung der linearen Regression**:
Es wurde gezeigt, wie man ein lineares Regressionsmodell mit `scikit-learn` erstellt, um eine Beziehung zwischen einer unabhängigen (Körpergröße) und einer abhängigen Variable (Ringfingerumfang) zu modellieren. Die Regressionslinie hilft, Trends zu quantifizieren und Vorhersagen zu treffen.


---


3.  **Bewertung der Modellgüte**:
Die Ausgabe des Bestimmtheitsmaßes (R²) hat uns gezeigt, wie gut unser lineares Modell die Varianz in den Daten erklärt. Ein R²-Wert nahe 1 deutet auf eine gute Passform des Modells hin, was für die Interpretation und Vertrauenswürdigkeit des Modells entscheidend ist.
