## Abschnitt 2 – Listen und Schleifen

Listen sind geordnete Sammlungen von Werten.  
Mit einer Schleife können wir über alle Elemente einer Liste nacheinander iterieren.


```python
zahlen = [2, 4, 6, 8, 10]
print("Liste:", zahlen)
print("Erstes Element:", zahlen[0])
print("Anzahl Elemente:", len(zahlen))
```

```python
for z in zahlen:
    print("Quadrat von", z, "=", z**2)
```

## Abschnitt 4 – Eine kleine Tabelle mit pandas

Ein **DataFrame** ist eine tabellarische Datenstruktur mit Zeilen und Spalten.  
So ähnlich werden später auch viele Datensätze im Machine Learning dargestellt.


```python
df = pd.DataFrame({
    "Lernzeit_Stunden": [2, 4, 6, 8, 10],
    "Punktzahl": [50, 58, 68, 80, 92]
})

df
```

```python
print("Form des DataFrames:", df.shape)
print()
print("Spaltennamen:", list(df.columns))
print()
print("Erste Zeilen:")
print(df.head())
```

## Optional: Interaktive Tabellenansicht in Google Colab

Diese Zelle funktioniert **nur in Google Colab**.  
In Jupyter wird stattdessen einfach das normale DataFrame angezeigt.


```python
try:
    from google.colab import sheets
    sheets.InteractiveSheet(df=df)
except ImportError:
    print("Interaktive Tabellenansicht ist nur in Google Colab verfügbar.")
    display(df)
```

## Abschnitt 5 – Erste Visualisierung

Wir erstellen ein einfaches Liniendiagramm.  
Später werden wir im Kurs häufiger Streudiagramme verwenden.


```python
plt.figure(figsize=(6, 4))
plt.plot(df["Lernzeit_Stunden"], df["Punktzahl"], marker="o")
plt.xlabel("Lernzeit in Stunden")
plt.ylabel("Punktzahl")
plt.title("Lernzeit und Punktzahl")
plt.grid(True)
plt.show()
```
